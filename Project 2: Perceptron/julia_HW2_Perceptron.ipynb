```julia
using PyPlot
N=100;# number of input vectors
# Define the symbols and colors we'll use in the plots later
symbols = ["ko","kx"];
mycolors = [0.5 0.5 1; 1 0.5 0.5];
classvals = [-1 1];

## generate random (linarly separable) data
trainPoints=randn(N,2).*1.5;
# defining random hyperplane
w=rand(2,1);
# assigning labels +1, -1 labels depending on what side of the plane they lie on
trainLabels=sign(trainPoints*w);
i=randperm(N);
# shuffling training points in random order
trainPoints=trainPoints[i,:]
trainLabels=trainLabels[i];

figure("Linear Classifier Visualization");
clf();
xkcd(false);
# call perceptron to find w from data
w=perceptron(trainPoints,trainLabels);
b=0;
# Generate the plot

res=300;
xrange = linspace(-5, 5,res);
yrange = linspace(-5,5,res);
pixelX=repmat(xrange',length(yrange),1);
pixelY=repmat(yrange, 1,length(xrange));
testPoints = [pixelX[:] pixelY[:] ones(length(pixelX[:]),1)];
testLabels=testPoints*[w b]';

Z = zeros(res,res);
Z[:]+=testLabels[:];
Z[1,1]=1; # this is to scale the colors correctly
contourf(pixelX, pixelY, sign(Z));
ColorMap(mycolors);
hold(true);
 for c = 1:2
    idx = find(trainLabels.==classvals[c]);
    hp=plot(trainPoints[idx,1], trainPoints[idx,2], symbols[c]);
 end
 quiver(0,0,w[1],w[2],linewidth=2,color=[0,1,0]);
 axis("equal");
grid(false);
 xticks(-5:5);
 yticks(-5:5);
 axis("tight");
 axis([-5,5,-5,5]);
```

<p> You can use the following script to visualize the weight vector and classification error on the digits data set. Your perceptron should be able to find a separating hyperplane for this data. </p>

```julia
using MAT
using PyPlot
function loaddata(filename)
    data=matread(filename)
    xTr=data["xTr"]'; # load in Training data
    yTr=round(Int,data["yTr"])'; # load in Training labels
    xTe=data["xTe"]'; # load in Testing data
    yTe=round(Int,data["yTe"])'; # load in Testing labels
    return xTr,yTr,xTe,yTe
end;

## load data
xTr,yTr,xTe,yTe=loaddata("../resource/lib/digits.mat");
MAXITER=10;
N=100;
cl=[0,7];
ii=find((yTr.==cl[2]) | (yTr.==cl[1]));
ii=ii[randperm(length(ii))];
ii=ii[1:N];
xTr=xTr[ii,:];
yTr=yTr[ii];
yTr=(yTr.==cl[1]).*2-1;


figure(1);
n=length(yTr);
set_cmap("gray");
## do loop
w=xTr[1,:].*0;
err=1.0;
for iter=1:MAXITER # run at moast MAXITER iterations
    for i=1:n
         # draw offender
         subplot(1,2,2);
         pcolormesh(flipdim(reshape(xTr[i,:],16,16),1));
         xticks([]);
         yticks([]);
         #xlabel([sprintf('Label: #i Pred:#2.2f',yTr[i],w'*xTr[i,:]]);
         title("Current Sample");

         if classifyLinear(xTr[i,:]',w[:]',0)!=yTr[i]
             ## do update
             w=perceptronUpdate(xTr[i,:]',yTr[i],w[:]');
             ## compute new training error
             preds=classifyLinear(xTr,w[:]',0);
             err=sum(preds.!=yTr)./length(yTr);

             ## plot new vector
             subplot(1,2,1);
            pcolormesh(flipdim(reshape(w,16,16),1));

             xlabel("Error $err");
             title("Weight Vector");
             xticks([]);
             yticks([]);
             #input ("Press a key>");             
        end;
        if err==0;
          break;
        end;
      end;
      print("Error: $err            \n" );
    pause(0.01);
    if err==0
        break;
    end;
end;
```
